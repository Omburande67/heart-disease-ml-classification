```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import xgboost as xgb
from imblearn.over_sampling import SMOTE
import pickle as pk

# Load data
data = pd.read_csv('heart_disease.csv')
data.drop(columns=['education'], inplace=True)
data.dropna(inplace=True)

# Encode categorical variables
data['Gender'] = data['Gender'].replace(['Male', 'Female'], [1, 0])
data['Heart_ stroke'] = data['Heart_ stroke'].replace(['No', 'yes'], [0, 1])
data['prevalentStroke'] = data['prevalentStroke'].replace(['no', 'yes'], [0, 1])

# Prepare data for training
data_input = data.drop(columns=['Heart_ stroke'])
data_output = data['Heart_ stroke']

# Apply SMOTE to handle class imbalance
sm = SMOTE(random_state=42)
data_input_resampled, data_output_resampled = sm.fit_resample(data_input, data_output)

# Split data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(data_input_resampled, data_output_resampled, test_size=0.2, random_state=42)

# Logistic Regression Model
logistic_model = LogisticRegression()
logistic_model.fit(x_train, y_train)
logistic_predictions = logistic_model.predict(x_test)
logistic_score = accuracy_score(y_test, logistic_predictions)
logistic_precision = precision_score(y_test, logistic_predictions)
logistic_recall = recall_score(y_test, logistic_predictions)
logistic_f1 = f1_score(y_test, logistic_predictions)

# Random Forest Model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(x_train, y_train)
rf_predictions = rf_model.predict(x_test)
rf_score = accuracy_score(y_test, rf_predictions)
rf_precision = precision_score(y_test, rf_predictions)
rf_recall = recall_score(y_test, rf_predictions)
rf_f1 = f1_score(y_test, rf_predictions)

# Bagging Classifier Model
bagging_model = BaggingClassifier(random_state=42)
bagging_model.fit(x_train, y_train)
bagging_predictions = bagging_model.predict(x_test)
bagging_score = accuracy_score(y_test, bagging_predictions)
bagging_precision = precision_score(y_test, bagging_predictions)
bagging_recall = recall_score(y_test, bagging_predictions)
bagging_f1 = f1_score(y_test, bagging_predictions)

# XGBoost Model
xg_model = xgb.XGBClassifier(random_state=42)
xg_model.fit(x_train, y_train)
xg_predictions = xg_model.predict(x_test)
xg_score = accuracy_score(y_test, xg_predictions)
xg_precision = precision_score(y_test, xg_predictions)
xg_recall = recall_score(y_test, xg_predictions)
xg_f1 = f1_score(y_test, xg_predictions)

# Save the models
pk.dump(logistic_model, open('Heart_disease_LR_model.pkl', 'wb'))
pk.dump(rf_model, open('Heart_disease_RF_model.pkl', 'wb'))
pk.dump(bagging_model, open('Heart_disease_Bagging_model.pkl', 'wb'))
pk.dump(xg_model, open('Heart_disease_XGBoost_model.pkl', 'wb'))

# Display the results
print("Logistic Regression Results:")
print(f"Accuracy: {logistic_score:.2f}")
print(f"Precision: {logistic_precision:.2f}")
print(f"Recall: {logistic_recall:.2f}")
print(f"F1 Score: {logistic_f1:.2f}")
print("\nRandom Forest Results:")
print(f"Accuracy: {rf_score:.2f}")
print(f"Precision: {rf_precision:.2f}")
print(f"Recall: {rf_recall:.2f}")
print(f"F1 Score: {rf_f1:.2f}")
print("\nBagging Classifier Results:")
print(f"Accuracy: {bagging_score:.2f}")
print(f"Precision: {bagging_precision:.2f}")
print(f"Recall: {bagging_recall:.2f}")
print(f"F1 Score: {bagging_f1:.2f}")
print("\nXGBoost Results:")
print(f"Accuracy: {xg_score:.2f}")
print(f"Precision: {xg_precision:.2f}")
print(f"Recall: {xg_recall:.2f}")
print(f"F1 Score: {xg_f1:.2f}")

# Example prediction with all models
input_sample = np.array([[0, 59, 0, 0, 0, 0, 1, 0, 209, 150, 85, 20.77, 90, 88]])

logistic_prediction = logistic_model.predict(input_sample)
rf_prediction = rf_model.predict(input_sample)
bagging_prediction = bagging_model.predict(input_sample)
xg_prediction = xg_model.predict(input_sample)

print("\nPrediction on Sample Input:")
print(f"Logistic Regression Prediction: {'Heart Disease' if logistic_prediction[0] == 1 else 'Healthy'}")
print(f"Random Forest Prediction: {'Heart Disease' if rf_prediction[0] == 1 else 'Healthy'}")
print(f"Bagging Classifier Prediction: {'Heart Disease' if bagging_prediction[0] == 1 else 'Healthy'}")
print(f"XGBoost Prediction: {'Heart Disease' if xg_prediction[0] == 1 else 'Healthy'}")
```
